## Using below given table answer questions a and b.

    (a) Using the dataset above, we want to build a decision tree which classifies Y as T/F given the binary variables A,B,C. Draw the tree that would be learned by the greedy algorithm with zero training error. You do not need to show any computation.

    (b) Is this tree optimal (i.e. does it get zero training error with minimal depth)? Explain in less than two sentences. If it is not optimal, draw the optimal tree as well.

```python
import pandas as pd
label_dict = {'A': ['F','T','T','T'],
              'B': ['F','F','T','T'],
              'C': ['F','T','F','T'],
              'Y':['F','T','T','F']}

df=pd.DataFrame(label_dict)
df.replace({'F' : 0.0, 'T' : 1.0}, inplace = True)
```

```python
Y=df.Y
df = df.drop('Y', axis = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

decision_tree = DecisionTreeClassifier(
    criterion='entropy',
    random_state = 1337
)


decision_tree.fit(df, Y)
```

```python
tree.plot_tree(decision_tree)
```

## We will use the dataset below to learn a decision tree which predicts if people pass machine learning (Yes or No), based on their previous GPA (High, Medium, or Low) and whether or not they studied.

    (a) What is the entropy H(Passed)?

    (b) What is the entropy H(Passed | GPA)?

    (c) What is the entropy H(Passed | Studied)?

```python
label_dict = {'GPA': ['L','L','M','M','H','H'],
            'Studies': ['F','T','F','T','F','T'],
            'Passed': ['F','T','F','T','T','T']}
df=pd.DataFrame(label_dict)
Y=df.Passed
df.replace({'L' : 2, 'M' : 3, 'H' : 4,'F' :0,'T' : 1}, inplace = True)
df = df.drop('Passed', axis = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree

decision_tree = DecisionTreeClassifier(
    criterion='entropy',
    random_state = 1337
)
```

```python
Y=Y.astype('int')
df=df.astype('int')
```

```python
decision_tree.fit(df, Y)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 10), dpi=200)
tree.plot_tree(decision_tree = decision_tree,
               feature_names=df.columns,
               filled = True
              )
plt.show()
```
